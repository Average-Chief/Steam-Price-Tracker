```python
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
import csv
import pandas
```

```python
url = 'https://store.steampowered.com/app/976730/Halo_The_Master_Chief_Collection/'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36 Edg/123.0.0.0"}
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content,"html")
print(soup)
```

```python
title = soup.find(id ="appHubAppName", class_="apphub_AppName").get_text().strip()
price = soup.find(class_ = "game_purchase_price price").get_text().strip()[2:]
reviews = soup.find(class_ ="game_review_summary positive").get_text().strip()
ratings = soup.find(class_="nonresponsive_hidden responsive_reviewdesc").get_text().strip()
requirements = soup.find(class_="bb_ul").get_text().strip()[:48]
today = datetime.date.today()
print(title, price, reviews, ratings, requirements, today, sep='\n')
```
